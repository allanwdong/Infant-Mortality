```python
import internal_ref as ref

import pandas as pd
import os.path
```

# Create Legend

```python
os.chdir(ref.Unemployment)
```

```python
series = ref.Unemployment + '/la.series.txt'
df_series = pd.read_csv(series, sep='\t', index_col=False)

df_series1 = df_series.drop(['area_type_code', 
                             'measure_code', 
                             'seasonal', 
                             'srd_code', 
                             'footnote_codes',
                             'begin_year', 
                             'begin_period', 
                             'end_year', 
                             'end_period'
                            ], axis =1)
```

```python
df_series1.head(3)
```

```python
areamaps = ref.Unemployment + '/la.areamaps.txt'
df_area = pd.read_csv(areamaps, sep='\t', index_col=False)

df_area1 = df_area.drop(['area_type_code',
                         'display_level',
                         'selectable',
                         'sort_sequence'
                        ], axis=1)
```

```python
df_area1.head(3)
```

```python
df_legend = pd.merge(df_series1, df_area1, how='left', left_on='area_code', right_on='area_code')
```

```python
df_legend.head(3)
```

```python
legend = ref.Unemployment + '/legend'
df_legend.to_csv(legend)
```

# Merge

```python
os.chdir(ref.Unemployment)
```

```python
_95_99 = ref.Unemployment + '/la.data.0.CurrentU95-99.txt'
df_9599 = pd.read_csv(_95_99, sep='\t')
```

```python
df_9599.head(3)
```

```python
_00_04 = ref.Unemployment + '/la.data.0.CurrentU00-04.txt'
df_0004 = pd.read_csv(_00_04, sep='\t')
```

```python
df_0004.head(3)
```

```python
# 05_09 file returned errors saying multiples dtypes, wanted specified
_05_09 = ref.Unemployment + '/la.data.0.CurrentU05-09.txt'
df_0509=pd.read_csv(_05_09, sep='\t', dtype={'series_id' : str, 
                                             'year' : int, 
                                             'period' : str, 
                                             'value' : object,
                                             'footnote_codes' : str})
```

```python
df_0509.head(3)
```

```python
_10_14 = ref.Unemployment + '/la.data.0.CurrentU10-14.txt'
df_1014 = pd.read_csv(_10_14, sep='\t')
```

```python
df_1014.head(3)
```

```python
df_99 = df_9599[df_9599.year==1999]
```

```python
df_99.head(3)
```

```python
df = pd.concat([df_99, df_0004, df_0509, df_1014], axis=0, ignore_index=True)
```

```python
df_merge = pd.merge(df, df_legend, how='left', left_on='series_id', right_on='series_id')
```

```python
df_unem = df_merge[df_merge.series_title.str.find(r'Unemployment Rate') != -1]
```

```python
df_unem.rename(columns = {'value':'unemployment_rate'}, inplace=True)
```

```python
df_unem.head(3)
```

```python
dfunemfilename = ref.Home + '/labor.csv'
df_unem.to_csv(dfunemfilename, index_label=False)
```
